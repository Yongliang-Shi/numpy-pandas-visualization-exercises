```python
import matplotlib.pyplot as plt

x = range(-5,5)
y = [i**2 - i + 2 for i in x]

plt.figure(figsize = (16, 9))

plt.scatter(x,y)
plt.plot(x,y)
plt.scatter(0,0)

plt.title(r"$y=x^2-x+2$")
plt.annotate('the orgin', xy=(0,0), xytext=(-0.3, 15), arrowprops={'arrowstyle': '->'})
```

    - square root: math.sqrt(x)
    - x**3
    - the tangent of x: math.tan(x)
    - 2 to the power of x: math.pow(2,x)

```python
import math

x1 = range(10)
y1 = [math.sqrt(i) for i in x1]

plt.scatter(x1,y1)
plt.plot(x1,y1)
plt.bar(x1,y1)
plt.title(r"$y = \sqrt{2}$")
```
