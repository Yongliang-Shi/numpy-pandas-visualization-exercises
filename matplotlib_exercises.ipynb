```python
import matplotlib.pyplot as plt

x = range(-5,5) # x = list(range(-5,5))
y = [i**2 - i + 2 for i in x] # generate y values

plt.figure(figsize = (16, 9))

plt.scatter(x,y)
plt.plot(x,y)
plt.scatter(0,0)

plt.title(r"$y=x^2-x+2$")
plt.annotate('the orgin', xy=(0,0), xytext=(-0.3, 15), arrowprops={'arrowstyle': '->'})
```

2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):
    - square root: math.sqrt(x)
    - x**3
    - the tangent of x: math.tan(x)
    - 2 to the power of x: math.pow(2,x)

```python
import math

# range of x: cann't be less then 0
x1 = range(10)

# generate y
y1 = [math.sqrt(i) for i in x1] # Review: for simplicity, sqrt() by itself i enough. 

plt.scatter(x1,y1)
plt.plot(x1,y1)
plt.bar(x1,y1)
plt.title(r"$y = \sqrt{x}$")

# Review: add x and y labels
```

```python
# range of x
x2 = range(10)

# generate y
y2 = [i**3 for i in x2]

plt.scatter(x2,y2)
plt.plot(x2,y2)
plt.bar(x2,y2)
plt.title(r"$y = x^3$")

# Review: add x and y labels
```

```python
x3 = range(10)
y3 = [math.tan(i) for i in x3]

plt.scatter(x3,y3)
plt.plot(x3,y3)
plt.bar(x3,y3)
plt.title(r"$y = tan(x)$")
```

```python
# range of x
x4 = range(10)

# generate y
y4 = [math.pow(2,i) for i in x4]

plt.scatter(x4,y4)
plt.plot(x4,y4)
plt.bar(x4,y4)
plt.title(r"$y = 2^x$")
```
