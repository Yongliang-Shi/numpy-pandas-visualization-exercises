```python
import matplotlib.pyplot as plt
import math

plt.figure(figsize=(8,4.5))

plt.title("Big O Notation")
plt.xlabel("Elements")
plt.ylabel("Operations")
```

```python
x = range(10)
y = [i + 1 for i in x]

plt.figure(figsize=(8,4.5))

plt.scatter(x,y)
plt.plot(x,y,label = "O1")
plt.legend()

plt.title("Big O Notation")
plt.xlabel("x")
plt.ylabel("y = x + 1")
```

```python
x = range(1, 10)
y = [math.log10(i) for i in x]

plt.figure(figsize=(8,4.5))

plt.scatter(x,y)
plt.plot(x,y,label = "O(log n)")
plt.legend()

plt.title("Big O Notation")
plt.xlabel("x")
plt.ylabel(r"$y = log(x)$")
```

```python
x = range(0, 10)
y = [i for i in x]

plt.figure(figsize=(8,4.5))

plt.scatter(x,y)
plt.plot(x,y,label = "O(n)")
plt.legend()

plt.title("Big O Notation")
plt.xlabel("x")
plt.ylabel("y = x")
```

```python
x = range(1, 10)
y = [i*math.log10(i) for i in x]

plt.figure(figsize=(8,4.5))

plt.scatter(x,y)
plt.plot(x,y,label = "O(n log n)")
plt.legend()

plt.title("Big O Notation")
plt.xlabel("x")
plt.ylabel(r"$y = x * log(x)$")
```

```python
x = range(0, 10)
y = [i**2 for i in x]

plt.figure(figsize=(8,4.5))

plt.scatter(x,y)
plt.plot(x,y,label = "O(n^2)")
plt.legend()

plt.title("Big O Notation")
plt.xlabel("x")
plt.ylabel(r"$y = x^2$")
```

```python
x = range(0, 10)
y = [math.pow(2,i) for i in x]

plt.figure(figsize=(8,4.5))

plt.scatter(x,y)
plt.plot(x,y,label = "O(2^n)")
plt.legend()

plt.title("Big O Notation")
plt.xlabel("x")
plt.ylabel(r"$y = 2^x$")
```

```python
x = range(0, 10)
y = [math.factorial(i) for i in x]

plt.figure(figsize=(8,4.5))

plt.scatter(x,y)
plt.plot(x,y,label = "O(n!)")
plt.legend()

plt.title("Big O Notation")
plt.xlabel("n")
plt.ylabel(r"$y = (n)!$")
```

```python
x = range(10)
y = [i**i for i in x]

plt.figure(figsize=(8,4.5))

plt.scatter(x,y)
plt.plot(x,y,label = "O(n^n)")
plt.legend()

plt.title("Big O Notation")
plt.xlabel("n")
plt.ylabel(r"$ y = n^n$")
```

```python
def horizontal_line(start, end, height):
    x = range(start,end)
    y = [height for i in x]
    return plt.plot(x,y)
    
horizontal_line(0,5,10) # valid
```

```python
def vertical_line(start, end, position):
    y = range(start, end)
    x = [position for i in y]
    return plt.plot(x,y)

vertical_line(1,10,5)
```

```python
def diagonal_line_positive(start, end):
    x = range(start, end)
    y = x
    return plt.plot(x,y)

diagonal_line_positive(0,5)
```

```python
def diagonal_line_negative(start, end):
    x = range(start,end)
    y = [-i for i in x]
    return plt.plot(x,y)

diagonal_line_negative(0,5)
```
